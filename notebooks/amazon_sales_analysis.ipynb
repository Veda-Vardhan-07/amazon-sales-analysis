## KPI Summary

- Calculated total sales and total profit.
- Counted the total number of unique orders.
- Computed the average sales and average profit per order.
- These KPIs provide an overview of the company's business performance.

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(category_sales)
```

```python
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.savefig("../images/category_sales.png", dpi=300, bbox_inches="tight")

plt.show()
```

## Category Analysis

- Calculated total sales for each product category.
- Technology generated the highest sales.
- Office Supplies generated the lowest sales.
- A bar chart was created to compare category performance.

```python
sub_category_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)
print(sub_category_sales)
```

```python
sub_category_sales.plot(kind="bar", figsize=(10,6))

plt.title("Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
top_products.plot(kind="barh", figsize=(10,6))

plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")

plt.show()
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month_name()
```

```python
df[["Order Date", "Year"]].head()
```

```python
df["Month No"] = df["Order Date"].dt.month
print(df["Month No"].head())
```

```python
monthly_sales = df.groupby(["Month No","Month"])["Sales"].sum().sort_index()
monthly_sales = monthly_sales.droplevel(0)
print(monthly_sales)
```

```python
monthly_sales.plot(kind="bar")

plt.title("Month-wise Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.savefig("../images/monthly_sales.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
yearly_sales = df.groupby("Year")["Sales"].sum()
print(yearly_sales)
```

```python
yearly_sales.plot(kind="bar")

plt.title("Year-wise Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")

plt.show()
```

## Business Insight

- Sales increased steadily from 2014 to 2017.
- 2017 recorded the highest total sales.
- The upward trend indicates business growth over the four-year period.

```python
regional_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print(regional_sales)
```

```python
regional_sales.plot(kind="bar")

plt.title("Region-wise Sales")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.savefig("../images/region_sales.png", dpi=300, bbox_inches="tight")

plt.show()
```

## Business Insight

- The West region generated the highest sales.

```python
region_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)
print(region_profit.map(lambda x: f"{x:,.2f}"))
```

```python
region_profit.plot(kind="bar")

plt.title("Region-wise Profit")
plt.xlabel("Region")
plt.ylabel("Profit")

plt.savefig("../images/region_profit.png", dpi=300, bbox_inches="tight")


plt.show()
```

## Business Insight

- The West region generated the highest overall profit.
- The Central region earned lower profit despite moderate sales.
- High sales do not always translate into high profit.

```python
discount_profit = df.groupby("Discount")["Profit"].sum().sort_index()
print(discount_profit)
```

```python
discount_profit.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Profit by Discount")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.savefig("../images/discount_profit.png", dpi=300, bbox_inches="tight")

plt.show()
```

## Business Insight

- Orders with little or no discount generated the highest profit.
- As discount levels increased, overall profit decreased.
- Some high-discount levels resulted in negative profit, indicating that excessive discounts reduce profitability.
